Mas como eu calculo a e b da eq da reta de modo que n√£o seja no chute? :

#Coeficiente angular


a = coeficiente angular
n = numero de elementos
y = moscow
x = datas
O coeficiente angular pode ser obtido usando a equa√ß√£o:

Capturar.PNG

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
```

```python
a
```

#Coeficiente linear (b)
b = media(Yi) - √¢ * m√©dia (Xi)

> Bloco com recuo



```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x,y)
plt.plot(41.5, 41.5*a+b, '*r')
plt.plot(100, 100*a+b, '*r')
```
